```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(16, 18))
fig.suptitle('Customer Churn Analysis Dashboard', fontsize=20)

# Plot 1: Churn Reasons
churned_customers = df[df['Churn Label'] == 'Yes']
churn_reason_counts = churned_customers['Churn Reason'].value_counts()
sns.barplot(x=churn_reason_counts.index, y=churn_reason_counts.values, ax=axes[0, 0])
axes[0, 0].set_xlabel('Churn Reason')
axes[0, 0].set_ylabel('Number of Customers')
axes[0, 0].set_title('Distribution of Churn Reasons')
axes[0, 0].tick_params(axis='x', rotation=90)

# Plot 2: Churn by Contract Type
churned_customers = df[df['Churn Value'] == 1]
contract_churn_counts = churned_customers['Contract'].value_counts()
sns.barplot(x=contract_churn_counts.index, y=contract_churn_counts.values, ax=axes[0, 1])
axes[0, 1].set_xlabel('Contract Type')
axes[0, 1].set_ylabel('Number of Churned Customers')
axes[0, 1].set_title('Churn Count by Contract Type')

# Plot 3: Churn by Monthly Charges
churned_monthly_charges = df[df['Churn Value'] == 1]['Monthly Charges']
non_churned_monthly_charges = df[df['Churn Value'] == 0]['Monthly Charges']
sns.histplot(churned_monthly_charges, color='skyblue', label='Churned', kde=True, ax=axes[1, 0])
sns.histplot(non_churned_monthly_charges, color='salmon', label='Non-Churned', kde=True, ax=axes[1, 0])
axes[1, 0].set_xlabel('Monthly Charges')
axes[1, 0].set_ylabel('Frequency')
axes[1, 0].set_title('Distribution of Monthly Charges for Churned vs. Non-Churned Customers')
axes[1, 0].legend()

# Plot 4: Churn by Tenure
sns.histplot(df[df['Churn Value'] == 1]['Tenure Months'], color='skyblue', label='Churned', kde=True, ax=axes[1, 1])
sns.histplot(df[df['Churn Value'] == 0]['Tenure Months'], color='salmon', label='Non-Churned', kde=True, ax=axes[1, 1])
axes[1, 1].set_xlabel('Tenure in Months')
axes[1, 1].set_ylabel('Frequency')
axes[1, 1].set_title('Distribution of Tenure Months for Churned vs. Non-Churned Customers')
axes[1, 1].legend()

# Plot 5: Churn by Payment Method
churned_customers = df[df['Churn Value'] == 1]
payment_method_churn_counts = churned_customers['Payment Method'].value_counts()
sns.barplot(x=payment_method_churn_counts.index, y=payment_method_churn_counts.values, ax=axes[2, 0])
axes[2, 0].set_xlabel('Payment Method')
axes[2, 0].set_ylabel('Number of Churned Customers')
axes[2, 0].set_title('Churn Count by Payment Method')
axes[2, 0].tick_params(axis='x', rotation=45)


# Plot 6: Churn by Internet Service Type
churned_customers = df[df['Churn Value'] == 1]
internet_service_churn_counts = churned_customers['Internet Service'].value_counts()
sns.barplot(x=internet_service_churn_counts.index, y=internet_service_churn_counts.values, ax=axes[2, 1])
axes[2, 1].set_xlabel('Internet Service Type')
axes[2, 1].set_ylabel('Number of Churned Customers')
axes[2, 1].set_title('Churn Count by Internet Service Type')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```
